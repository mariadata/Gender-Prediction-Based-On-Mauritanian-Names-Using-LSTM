```python
import pandas as pd 
import re
```

```python
#Load data 
df_bac = pd.read_excel('res-bac-2022.xlsx')
```

```python
#Extract first names
list_firstnames = []
for i in range (len(df_bac)) : 
    s = re.search('([^\s]+)',df_bac['Nom'][i])
    list_firstnames.append(s.group(0))
```

```python
list_firstnames[:5]
```

```python
#Add list_firstnames to df 
df_bac['prenom'] = list_firstnames
```

```python
df_bac.head(2)
```

```python
#function to preprocess names 
def preprocess(df):
    # Step 1: Lowercase
    df['prenom'] = df['prenom'].str.lower()

    # Step 2: Split individual characters
    df['prenom'] = [list(prenom) for prenom in df['prenom']]


    # Step 3: Pad names with spaces to make all names same length
    name_length = 50
    df['prenom'] = [
        (prenom + [' ']*name_length)[:name_length] 
        for prenom in df['prenom']
    ]

    # Step 4: Encode Characters to Numbers
    df['prenom'] = [
        [
            max(0.0, ord(char)-96.0) 
            for char in prenom
        ]
        for prenom in df['prenom']
    ]
    
    
    return df
```

```python
preprocess(df_bac)
```

```python
from tensorflow.keras.models import load_model
import numpy as np
```

```python
# Predictions
pred_model = load_model('MouF_final2.h5')

result = pred_model.predict(np.asarray(
    df_bac['prenom'].values.tolist())).squeeze(axis=1)

df_bac['M ou F?'] = [
    'M' if logit > 0.5 else 'F' for logit in result
]

df_bac['Probabilité'] = [
    logit if logit > 0.5 else 1.0 - logit for logit in result
]


df_bac['Probabilité'] = df_bac['Probabilité'].round(2)
df_bac = df_bac.drop(['prenom'], axis = 1)

```
